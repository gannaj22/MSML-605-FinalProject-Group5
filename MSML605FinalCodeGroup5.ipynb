```python
!pip install earthaccess xarray numpy matplotlib pyproj

import earthaccess
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import os
import pyproj
import calendar
# Setup
ECCO_DIR = "./data"

# Earthdata login
auth = earthaccess.login(strategy="interactive")

# ---------- DOWNLOAD BEDMACHINE ----------
print("🔍 Searching BedMachine Antarctica v3...")
bedmachine_result = earthaccess.search_data(
    short_name="NSIDC-0756",
    cloud_hosted=False,
    count=1
)
earthaccess.download(bedmachine_result, "./")

# ---------- FIND BEDMACHINE FILE ----------
for f in os.listdir("./"):
    if f.endswith(".nc") and "BedMachine" in f:
        BEDMACHINE_FILE = f
        break

if BEDMACHINE_FILE is None:
    raise FileNotFoundError("BedMachine file not found!")

# ---------- LOAD BEDMACHINE----------
ds_bm = xr.open_dataset(BEDMACHINE_FILE)

# ---------- LOAD ECCO----------
os.makedirs(ECCO_DIR, exist_ok=True)
# Years and months range
years = range(2014, 2017)  # 2016 to 2018 inclusive
months = range(1, 13)

# Loop through each month
for year in years:
    for month in months:
        # Format start and end dates for the month
        start_date = f"{year}-{month:02d}-01"
        end_day = calendar.monthrange(year, month)[1]
        end_date = f"{year}-{month:02d}-{end_day}"

        print(f"🔍 Searching: {start_date} to {end_date}")

        # Search for ECCO monthly climatology
        results = earthaccess.search_data(
            short_name="ECCO_L4_TEMP_SALINITY_05DEG_MONTHLY_V4R4",
            temporal=(start_date, end_date),
            bounding_box=(-180, -90, 180, 90),
        )

        # Download to ECCO_DIR
        try:
            earthaccess.download(results, ECCO_DIR)
        except Exception as e:
            print(f"Failed to download {start_date} → {e}")

print("All downloads complete.")

import numpy as np
import xarray as xr
import pyproj
from scipy.interpolate import RegularGridInterpolator
import os
import matplotlib.pyplot as plt

# ==== Step 1: Setup Paths and List ECCO Files ====
ECCO_DIR = "./data"  # Update to your ECCO directory
ecco_nc_files = [f for f in os.listdir(ECCO_DIR) if f.endswith(".nc")]
print("Found ECCO files:", ecco_nc_files)

ds_bm = xr.open_dataset(BEDMACHINE_FILE)
bed = ds_bm["bed"]
thickness = ds_bm["thickness"]  # Ice thickness variable from BedMachine
x_bm = ds_bm["x"]
y_bm = ds_bm["y"]

# Define region of interest (adjust as needed)
x_min, x_max = -2100000, -1900000
y_min, y_max = -600000, -400000
bed_clip = bed.sel(x=slice(x_min, x_max), y=slice(y_max, y_min))  # Y reversed
thickness_clip = thickness.sel(x=slice(x_min, x_max), y=slice(y_max, y_min))

# Calculate ice draft (depth of ice base below sea level, negative values)
ice_draft_clip = bed_clip + thickness_clip  # ice draft = bed + thickness

# Create meshgrid of BedMachine coordinates
x_bed = bed_clip['x'].values
y_bed = bed_clip['y'].values
xx_bed, yy_bed = np.meshgrid(x_bed, y_bed)

# Convert BedMachine x/y to lat/lon
transformer = pyproj.Transformer.from_crs("EPSG:3031", "EPSG:4326", always_xy=True)
lon_bed, lat_bed = transformer.transform(xx_bed, yy_bed)
patch_size = 64
stride = 16
patches = []

for ecco_file in ecco_nc_files:
    print(f"Processing {ecco_file}")
    ds_ecco = xr.open_dataset(os.path.join(ECCO_DIR, ecco_file))
    temp_ecco = ds_ecco["THETA"].isel(Z=10, time=0)
    salt_ecco = ds_ecco["SALT"].isel(Z=10, time=0)

    ecco_lats = temp_ecco.latitude.values
    ecco_lons = temp_ecco.longitude.values
    temp_ecco_data = temp_ecco.values
    salt_ecco_data = salt_ecco.values

    temp_interp_func = RegularGridInterpolator((ecco_lats, ecco_lons), temp_ecco_data, bounds_error=False, fill_value=np.nan)
    salt_interp_func = RegularGridInterpolator((ecco_lats, ecco_lons), salt_ecco_data, bounds_error=False, fill_value=np.nan)

    points = np.column_stack([lat_bed.ravel(), lon_bed.ravel()])
    temp_on_bedmachine = temp_interp_func(points).reshape(xx_bed.shape)
    salt_on_bedmachine = salt_interp_func(points).reshape(xx_bed.shape)

    # Extract patches including ice draft
    for i in range(0, bed_clip.shape[0] - patch_size + 1, stride):
        for j in range(0, bed_clip.shape[1] - patch_size + 1, stride):
            bed_patch = bed_clip.values[i:i+patch_size, j:j+patch_size]
            thickness_patch = thickness_clip.values[i:i+patch_size, j:j+patch_size]
            ice_draft_patch = ice_draft_clip.values[i:i+patch_size, j:j+patch_size]
            temp_patch = temp_on_bedmachine[i:i+patch_size, j:j+patch_size]
            salt_patch = salt_on_bedmachine[i:i+patch_size, j:j+patch_size]

            # Strictly skip any patch with any NaN
            if (np.isnan(bed_patch).any() or np.isnan(temp_patch).any() or
                np.isnan(salt_patch).any() or np.isnan(ice_draft_patch).any()):
                continue

            # Stack all variables: bed, ice draft, temperature, salinity
            patch = np.stack([bed_patch, ice_draft_patch, temp_patch, salt_patch], axis=-1)
            patches.append(patch)
            if len(patches) >= 11000:
                break
        if len(patches) >= 11000:
            break
    if len(patches) >= 11000:
        break

patches = np.array(patches)
print("Extracted patches:", patches.shape)
print("Any NaNs in patches?", np.isnan(patches).any())

# ==== Step 2: Compute Melt Rate Using Quadratic-Local Parameterization ====

# Extract variables from patches
bed = patches[..., 0]            # Bed elevation (meters, negative below sea level)
draft = patches[..., 1]          # Ice draft (meters, negative below sea level)
temp = patches[..., 2]           # Ocean temperature (deg C)
salt = patches[..., 3]           # Ocean salinity (psu)

# ISMIP6 freezing point constants
a = -0.0573
b = 0.0832
c = -7.53e-4

# Compute local freezing point at the ice base
T_f = a * salt + b + c * draft

# Compute thermal forcing (difference between ocean temp and local freezing point)
TF = temp - T_f

# Quadratic-local melt parameterization
gamma = 1.0  # m/yr/(degC)^2, tune as needed
melt_rate = gamma * TF**2

# Clip to physical bounds (0 to 50 m/yr typical for Antarctic ice shelves)
melt_rate = np.clip(melt_rate, 0, 50)

# Sanity check
print("Melt range (raw):", np.nanmin(melt_rate), np.nanmax(melt_rate))
print("Thermal forcing range:", np.nanmin(TF), np.nanmax(TF))
print("Any NaNs in melt_rate?", np.isnan(melt_rate).any())

# ==== Step 3: Save ====
np.save("meltnet_input_patches.npy", patches)
np.save("meltnet_melt_targets.npy", melt_rate)
print("Saved as meltnet_input_patches.npy and meltnet_melt_targets.npy")

# ==== Optional: Quick Visualization of a Patch ====
plt.figure(figsize=(15,4))
plt.subplot(1,4,1)
plt.imshow(patches[0][:,:,0], cmap='terrain')
plt.title('Bed Elevation')
plt.subplot(1,4,2)
plt.imshow(patches[0][:,:,1], cmap='Blues')
plt.title('Ice Draft')
plt.subplot(1,4,3)
plt.imshow(patches[0][:,:,2], cmap='plasma')
plt.title('Temperature')
plt.subplot(1,4,4)
import matplotlib.colors as colors
plt.imshow(melt_rate[0], cmap='inferno', norm=colors.LogNorm(vmin=1e-3, vmax=50))
plt.title('Melt Rate')
plt.tight_layout()
plt.show()

import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# =======================
# 1. Data Loading
# =======================
def load_meltnet_data():
    X = np.load("meltnet_input_patches.npy")  # (N, 64, 64, 4)
    Y = np.load("meltnet_melt_targets.npy")    # (N, 64, 64)
    print("Loaded X shape:", X.shape)
    print("Loaded Y shape:", Y.shape)
    means = X.mean(axis=(0,1,2), keepdims=True)
    stds = X.std(axis=(0,1,2), keepdims=True)
    X_norm = (X - means) / (stds + 1e-8)
    return X_norm.astype(np.float32), Y.astype(np.float32), means.squeeze(), stds.squeeze()

# =======================
# 2. MELTNET Model
# =======================
class Swish(nn.Module):
    def forward(self, x):
        return x * torch.sigmoid(x)

class SqueezeExcite(nn.Module):
    def __init__(self, channels, reduction=16):
        super().__init__()
        self.se = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(channels, channels//reduction, 1),
            Swish(),
            nn.Conv2d(channels//reduction, channels, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return x * self.se(x)

class SEGNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(4, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            Swish(),
            SqueezeExcite(64),
            nn.Conv2d(64, 128, 3, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            Swish()
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(128, 64, 3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(64),
            Swish(),
            nn.Conv2d(64, 128, 3, padding=1)
        )
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
    def forward(self, x):
        x_enc = self.encoder(x)
        x_dec = self.decoder(x_enc)
        return x_dec

class DAENetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(128, 64, 3, padding=1),
            Swish(),
            nn.Conv2d(64, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            Swish(),
            nn.Conv2d(32, 1, 3, padding=1)
        )
    def forward(self, x):
        return self.net(x)

class MeltNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.seg = SEGNetwork()
        self.dae = DAENetwork()
    def forward(self, x):
        x = self.seg(x)
        return self.dae(x)

# =======================
# 3. Training and Validation
# =======================
def calculate_metrics(y_true, y_pred):
    y_pred = np.clip(y_pred, 0, 50)
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    nrmse = rmse / 50  # normalize by max melt
    cc = np.corrcoef(y_true.flatten(), y_pred.flatten())[0,1]
    return nrmse, cc

def validate(model, X, Y, device, loss_fn, batch_size):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        for i in range(0, len(X), batch_size):
            xb = torch.tensor(X[i:i+batch_size]).permute(0,3,1,2).to(device)
            yb = torch.tensor(Y[i:i+batch_size]).unsqueeze(1).to(device)
            pred = model(xb)
            total_loss += loss_fn(pred, yb).item()
    return total_loss / (len(X)//batch_size)

def train_model(model, X_train, Y_train, X_val, Y_val, device, batch_size=32, epochs=5):
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-5)
    steps_per_epoch = (len(X_train) + batch_size - 1) // batch_size
    total_steps = epochs * steps_per_epoch
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer, max_lr=1e-3, total_steps=total_steps,
        epochs=epochs, steps_per_epoch=steps_per_epoch
    )
    loss_fn = nn.HuberLoss()
    best_loss = float('inf')

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0
        perm = torch.randperm(len(X_train))
        X_shuffled = X_train[perm]
        Y_shuffled = Y_train[perm]
        for i in range(0, len(X_train), batch_size):
            end_idx = min(i+batch_size, len(X_train))
            xb = torch.tensor(X_shuffled[i:end_idx]).permute(0,3,1,2).to(device)
            yb = torch.tensor(Y_shuffled[i:end_idx]).unsqueeze(1).to(device)
            optimizer.zero_grad()
            pred = model(xb)
            loss = loss_fn(pred, yb)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()
            epoch_loss += loss.item()
        val_loss = validate(model, X_val, Y_val, device, loss_fn, batch_size)
        print(f"Epoch {epoch+1}/{epochs} | Train Loss: {epoch_loss/len(X_train):.4f} | Val Loss: {val_loss:.4f}")
        if val_loss < best_loss:
            best_loss = val_loss
            torch.save(model.state_dict(), 'best_model.pth')
    print("Training complete.")


# =======================
# 4. Visualization
# =======================
def plot_meltnet_comparison(model, sample_input, true_melt, device='cpu'):
    model.eval()
    with torch.no_grad():
        input_tensor = torch.tensor(sample_input).permute(2,0,1).unsqueeze(0).to(device)
        pred_melt = model(input_tensor).squeeze().cpu().numpy()
    nrmse, cc = calculate_metrics(true_melt, pred_melt)
    fig, axs = plt.subplots(1, 6, figsize=(24,4))
    channels = ['Bed Elevation', 'Ice Draft', 'Temperature', 'Salinity']
    for i in range(4):
        axs[i].imshow(sample_input[:,:,i], cmap='viridis')
        axs[i].set_title(channels[i])
    axs[4].imshow(true_melt, cmap='inferno', vmin=0, vmax=50)
    axs[4].set_title('True Melt Rate')
    im = axs[5].imshow(pred_melt, cmap='inferno', vmin=0, vmax=50)
    axs[5].set_title(f'Predicted Melt Rate\nNRMSE: {nrmse:.3f}, CC: {cc:.3f}')
    plt.colorbar(im, ax=axs[5], fraction=0.046, pad=0.04)
    plt.tight_layout()
    plt.show()




if __name__ == "__main__":
    import time

    # Load and subset data
    X, Y, means, stds = load_meltnet_data()
    X_small = X[:1000]
    Y_small = Y[:1000]
    X_train, X_test, Y_train, Y_test = train_test_split(
        X_small, Y_small, test_size=0.2, random_state=42
    )

    batch_size = 8

    # =======================
    # Benchmark: Float32 on CPU
    # =======================
    device_cpu = torch.device("cpu")
    model_cpu_f32 = MeltNet().to(device_cpu)
    print("Starting float32 training on CPU...")
    start_time = time.time()
    train_model(model_cpu_f32, X_train, Y_train, X_test, Y_test, device_cpu, batch_size=batch_size, epochs=1)
    cpu_f32_time = time.time() - start_time
    print(f"CPU (float32) Training Time: {cpu_f32_time:.2f} seconds")

    # =======================
    # Benchmark: Float32 on GPU (if available)
    # =======================
    if torch.cuda.is_available():
        device_gpu = torch.device("cuda")
        model_gpu_f32 = MeltNet().to(device_gpu)
        print("Starting float32 training on GPU...")
        start_time = time.time()
        train_model(model_gpu_f32, X_train, Y_train, X_test, Y_test, device_gpu, batch_size=batch_size, epochs=1)
        gpu_f32_time = time.time() - start_time
        print(f"GPU (float32) Training Time: {gpu_f32_time:.2f} seconds")

    # =======================
    # Benchmark: Float64 on CPU
    # =======================
    X_train64 = X_train.astype(np.float64)
    X_test64 = X_test.astype(np.float64)
    Y_train64 = Y_train.astype(np.float64)
    Y_test64 = Y_test.astype(np.float64)

    model_cpu_f64 = MeltNet().double().to(device_cpu)
    print("Starting float64 training on CPU...")
    start_time = time.time()
    train_model(model_cpu_f64, X_train64, Y_train64, X_test64, Y_test64, device_cpu, batch_size=batch_size, epochs=1)
    cpu_f64_time = time.time() - start_time
    print(f"CPU (float64) Training Time: {cpu_f64_time:.2f} seconds")

    # =======================
    # Visualization of Best Model (Float32 GPU or CPU)
    # =======================
    device_final = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = MeltNet().to(device_final)
    model.load_state_dict(torch.load("best_model.pth"))

    # Choose most active test sample
    melt_sums = np.sum(Y_test, axis=(1, 2))
    best_idx = np.argmax(melt_sums)
    test_sample = X_test[best_idx], Y_test[best_idx]

    plot_meltnet_comparison(model, test_sample[0], test_sample[1], device=device_final)


import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# =======================
# 1. Data Loading
# =======================
def load_meltnet_data():
    X = np.load("meltnet_input_patches.npy")  # (N, 64, 64, 4)
    Y = np.load("meltnet_melt_targets.npy")    # (N, 64, 64)
    print("Loaded X shape:", X.shape)
    print("Loaded Y shape:", Y.shape)
    means = X.mean(axis=(0,1,2), keepdims=True)
    stds = X.std(axis=(0,1,2), keepdims=True)
    X_norm = (X - means) / (stds + 1e-8)
    return X_norm.astype(np.float32), Y.astype(np.float32), means.squeeze(), stds.squeeze()

# =======================
# 2. MELTNET Model
# =======================
class Swish(nn.Module):
    def forward(self, x):
        return x * torch.sigmoid(x)

class SqueezeExcite(nn.Module):
    def __init__(self, channels, reduction=16):
        super().__init__()
        self.se = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Conv2d(channels, channels//reduction, 1),
            Swish(),
            nn.Conv2d(channels//reduction, channels, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return x * self.se(x)

class SEGNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(4, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            Swish(),
            SqueezeExcite(64),
            nn.Conv2d(64, 128, 3, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(128),
            Swish()
        )
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(128, 64, 3, stride=2, padding=1, output_padding=1),
            nn.BatchNorm2d(64),
            Swish(),
            nn.Conv2d(64, 128, 3, padding=1)
        )
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
    def forward(self, x):
        x_enc = self.encoder(x)
        x_dec = self.decoder(x_enc)
        return x_dec

class DAENetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(128, 64, 3, padding=1),
            Swish(),
            nn.Conv2d(64, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            Swish(),
            nn.Conv2d(32, 1, 3, padding=1)
        )
    def forward(self, x):
        return self.net(x)

class MeltNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.seg = SEGNetwork()
        self.dae = DAENetwork()
    def forward(self, x):
        x = self.seg(x)
        return self.dae(x)

# =======================
# 3. Training and Validation
# =======================
def calculate_metrics(y_true, y_pred):
    y_pred = np.clip(y_pred, 0, 50)
    rmse = np.sqrt(np.mean((y_true - y_pred)**2))
    nrmse = rmse / 50  # normalize by max melt
    cc = np.corrcoef(y_true.flatten(), y_pred.flatten())[0,1]
    return nrmse, cc

def validate(model, X, Y, device, loss_fn, batch_size):
    model.eval()
    total_loss = 0
    with torch.no_grad():
        for i in range(0, len(X), batch_size):
            xb = torch.tensor(X[i:i+batch_size]).permute(0,3,1,2).to(device)
            yb = torch.tensor(Y[i:i+batch_size]).unsqueeze(1).to(device)
            pred = model(xb)
            total_loss += loss_fn(pred, yb).item()
    return total_loss / (len(X)//batch_size)

def train_model(model, X_train, Y_train, X_val, Y_val, device, batch_size=32, epochs=5):
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=1e-5)
    steps_per_epoch = (len(X_train) + batch_size - 1) // batch_size
    total_steps = epochs * steps_per_epoch
    scheduler = torch.optim.lr_scheduler.OneCycleLR(
        optimizer, max_lr=1e-3, total_steps=total_steps,
        epochs=epochs, steps_per_epoch=steps_per_epoch
    )
    loss_fn = nn.HuberLoss()
    best_loss = float('inf')

    for epoch in range(epochs):
        model.train()
        epoch_loss = 0
        perm = torch.randperm(len(X_train))
        X_shuffled = X_train[perm]
        Y_shuffled = Y_train[perm]
        for i in range(0, len(X_train), batch_size):
            end_idx = min(i+batch_size, len(X_train))
            xb = torch.tensor(X_shuffled[i:end_idx]).permute(0,3,1,2).to(device)
            yb = torch.tensor(Y_shuffled[i:end_idx]).unsqueeze(1).to(device)
            optimizer.zero_grad()
            pred = model(xb)
            loss = loss_fn(pred, yb)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
            scheduler.step()
            epoch_loss += loss.item()
        val_loss = validate(model, X_val, Y_val, device, loss_fn, batch_size)
        print(f"Epoch {epoch+1}/{epochs} | Train Loss: {epoch_loss/len(X_train):.4f} | Val Loss: {val_loss:.4f}")
        if val_loss < best_loss:
            best_loss = val_loss
            torch.save(model.state_dict(), 'best_model.pth')
    print("Training complete.")

# =======================
# 4. Visualization
# =======================
def plot_meltnet_comparison(model, sample_input, true_melt, device='cpu'):
    model.eval()
    with torch.no_grad():
        input_tensor = torch.tensor(sample_input).permute(2,0,1).unsqueeze(0).to(device)
        pred_melt = model(input_tensor).squeeze().cpu().numpy()
    nrmse, cc = calculate_metrics(true_melt, pred_melt)
    fig, axs = plt.subplots(1, 6, figsize=(24,4))
    channels = ['Bed Elevation', 'Ice Draft', 'Temperature', 'Salinity']
    for i in range(4):
        axs[i].imshow(sample_input[:,:,i], cmap='viridis')
        axs[i].set_title(channels[i])
    axs[4].imshow(true_melt, cmap='inferno', vmin=0, vmax=50)
    axs[4].set_title('True Melt Rate')
    im = axs[5].imshow(pred_melt, cmap='inferno', vmin=0, vmax=50)
    axs[5].set_title(f'Predicted Melt Rate\nNRMSE: {nrmse:.3f}, CC: {cc:.3f}')
    plt.colorbar(im, ax=axs[5], fraction=0.046, pad=0.04)
    plt.tight_layout()
    plt.show()
```
